Turning a dictionary to a csv

```python
states = ['California','Texas','Florida','New York']
population = [12341354,12354125,12451251,145435234]

dict_states = {'States': states,'Population': population}

df_states = pd.DataFrame.from_dict(dict_states)        #this command means that we want to create a
                                                          # dataframe from a dictionary which is inside the parenthesis
print(df_states)

df_states.to_csv('states.csv', index=False)  #command for converting to csv file also with no indexing
```

Handling exception errors: Try-except

```python
new_list = [4,8,12,'California']

for element in new_list:
  try:
    print(element/2)
  except:
    print('The element is not a number!')      #this function avoids errors containing different type of file types
```

Getting html from a website

```python
from bs4 import BeautifulSoup
import requests

website = 'https://subslikescript.com/movie/Ludo-7212754'
result = requests.get(website)
content = result.text

soup = BeautifulSoup(content,'lxml')
print(soup.prettify())
```

Scraping a single page

```python
box = soup.find('article', class_='main-article')
title = box.find('h1').get_text()
transcript = box.find('div',class_='full-script').get_text(strip = True, separator = ' ')

with open(f'{title}.txt','w') as file:
  file.write(transcript)
```

Scraping Multiple Links within The Same Page

```python
root = 'https://subslikescript.com'
website = f'{root}/movies'
result = requests.get(website)
content = result.text
soup = BeautifulSoup(content,'lxml')

box = soup.find('article', class_='main-article')

links = []
for link in box.find_all('a', href=True):
  links.append(link['href'])

for link in links:
  try:
    print(link)
    website = f'{root}/{link}'
    result = requests.get(website)
    content = result.text
    soup = BeautifulSoup(content,'lxml')

    box = soup.find('article', class_='main-article')

    title = box.find('h1').get_text()
    transcript = box.find('div',class_='full-script').get_text(strip = True, separator = ' ')

    with open(f'{title}.txt','w') as file:
      file.write(transcript)
  except:
      print('------link not working------')
      print(link)
```
